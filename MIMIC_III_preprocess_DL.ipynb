```python
def pivot(df):
    df['CHARTTIME'] = pd.to_datetime(df['CHARTTIME'])
    # df.drop(columns=['ROW_ID','STORETIME', 'CGID','WARNING', 'ERROR', 'RESULTSTATUS', 'STOPPED'])

    pivoted_df = df.pivot_table(index=['ICUSTAY_ID','HADM_ID', 'SUBJECT_ID', 'CHARTTIME'], columns='ITEMID', values='VALUE', aggfunc='first')
    pivoted_df.reset_index(inplace=True)
    return pivoted_df
```

```python
def filter_first_3_hours(group):
    min_time = group['CHARTTIME'].min()
    time_threshold = min_time + pd.Timedelta(hours=3)
    return group[group['CHARTTIME'] <= time_threshold]
```

```python
def adjust_dataframe(df, target_columns, essential_columns, dtype_dict):
    # Ensure target_columns is a list
    # if not isinstance(target_columns, list):
    #     target_columns = [target_columns]

    # Add target columns to desired_columns list if they are not already there
    desired_columns = list(set(target_columns + essential_columns))

    # Add each target column to the DataFrame if it doesn't exist
    for col in target_columns:
        if col not in df.columns:
            df[col] = np.nan

    # Drop columns not in the desired columns list, while preserving essential columns
    columns_to_drop = set(df.columns) - set(desired_columns)
    df.drop(columns=columns_to_drop, inplace=True, errors='ignore')

    # Set the data type for each column based on the dtype_dict
    for col in df.columns:
        if col in dtype_dict:
            df[col] = df[col].astype(dtype_dict[col])

    grouped = df.groupby('ICUSTAY_ID').apply(filter_first_3_hours)
    # assert len(grouped.columns) == 30
    return grouped
```

```python
def aggregate(df, group, identifiers=['ICUSTAY_ID','HADM_ID', 'SUBJECT_ID']):
    # cat = list(df.drop(columns=identifiers).select_dtypes(include=['object']).columns)
    # num = list(df.drop(columns=identifiers).select_dtypes(include=['float64', 'int64']).columns)
    # assert set(cat)+set(num) == set(features)
    aggregations = {}
    for col, dtype in df.drop(columns=identifiers).dtypes.items():
        if dtype in ['int64', 'float64']:  # Replace with appropriate numeric types
            aggregations[col] = 'mean'
        elif dtype == 'object':  # Assuming 'object' dtype are categorical
            aggregations[col] = 'first'
        # Add more conditions if needed

    agged = df.groupby(group).agg(aggregations).reset_index()

    assert len(agged.columns) == 30
    return agged
```

```python
def add_features(df, features):

    if 'ADMISSIONS' in features:
        ref = pd.read_csv(os.path.join(datadir, tables['ADMISSIONS']))
        df['MORTALITY'] = df['HADM_ID'].isin(ref['HADM_ID'][pd.notna(ref['DEATHTIME'])])
        df['MORTALITY'] = df['MORTALITY'].astype('bool')
        df = df.merge(ref[['HADM_ID', 'DIAGNOSIS']], on='HADM_ID', how='left')

    if 'PATIENTS' in features and 'ICUSTAYS' in features:
        ref1 = pd.read_csv(os.path.join(datadir, tables['PATIENTS']))
        ref2 = pd.read_csv(os.path.join(datadir, tables['ICUSTAYS']))

        df = df.merge(ref1[['DOB', 'SUBJECT_ID']], on='SUBJECT_ID', how='left')
        df = df.merge(ref2[['ICUSTAY_ID', 'OUTTIME']], on='ICUSTAY_ID', how='left')
        df['AGE'] = pd.to_datetime(df['OUTTIME']).dt.year - pd.to_datetime(df['DOB']).dt.year
        df.loc[df['AGE'] >= 300, 'AGE'] = df['AGE'] - 250
        df['AGE'] = df['AGE'].astype('float64')

        df.drop(columns = ['DOB', 'OUTTIME'], inplace=True)

    return df
```

```python
def fill_nan(df):
    original = df.columns
    fix = ['HADM_ID', 'ICUSTAY_ID', 'SUBJECT_ID', 'AGE', 'MORTALITY', 'DIAGNOSIS', 'CHARTTIME', 'IS_DUMMY_ROW']
    for col in df.columns:
        if col in fix:
            continue
        df["HAS_"+str(col)] = np.nan
        df["HAS_"+str(col)] = df["HAS_"+str(col)].astype('bool')

    # Add indicators
    for i in range(len(df)):
        for col in original:
            if col in fix:
                continue
            else:
                if pd.notna(df.loc[i, col]):
                    df.loc[i, "HAS_"+str(col)] = True
                    continue
                else:
                    df.loc[i, "HAS_"+str(col)] = False
                    df.loc[i, col]= 'None' if df[col].dtypes == 'object' else 0


    return df
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)



newfile = "DATASET_GRU.csv"
newdf = pd.DataFrame()
if os.path.exists(os.path.join(datadir, newfile)):
    os.remove(os.path.join(datadir, newfile))

targets = list(dtypes.keys())[3:]
essentials = ['ICUSTAY_ID', 'HADM_ID', 'AGE', 'MORTALITY', 'SUBJECT_ID', 'DIAGNOSIS', 'CHARTTIME']
for i, chunk in enumerate(pd.read_csv(os.path.join(datadir, tables['CHARTEVENTS']), chunksize=10**6)):
    df = chunk[chunk['ICUSTAY_ID'].isin(icustay_id['ICUSTAY_ID'])]
    df = df[df['VALUE'].notna()]
    df = pivot(df)
    df = adjust_dataframe(df, list(targets), essentials, dtypes)
    # df = aggregate(df, ['ICUSTAY_ID', 'HADM_ID', 'SUBJECT_ID'])
    df = add_features(df, ['PATIENTS', 'ICUSTAYS', 'ADMISSIONS'])
    # grouped = df.groupby('ICUSTAY_ID')
    # df = pd.concat([process_patient(group) for name, group in grouped]).reset_index(drop=True)
    # df['IS_DUMMY_ROW'] = df['IS_DUMMY_ROW'].fillna(False).astype(bool)
    df = fill_nan(df)

    df['MORTALITY'] = df['MORTALITY'].astype('bool')
    df['HADM_ID'], df['SUBJECT_ID'] = df['HADM_ID'].astype('float64'), df['SUBJECT_ID'].astype('float64')

    if i!=0:
        if not newdf.columns.equals(df.columns):
            df = df.reindex(columns=newdf.columns)
        assert newdf.dtypes.equals(df.dtypes), "dtypes do not match"


    newdf = df

    with open(os.path.join(datadir, newfile), 'a') as f:
        df.to_csv(f, header=f.tell()==0, index=False)


    print(f"Saved {i}th chunk")
```
